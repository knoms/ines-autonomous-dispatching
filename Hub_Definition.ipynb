# Importing Libraries

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

# Loading Data from CSV files

```python
# get the stores
stores = pd.read_csv("coordinates_stores.csv")

# get the customers
customers20 = pd.read_csv("customers20.csv")
customers30 = pd.read_csv("customers30.csv")
customers40 = pd.read_csv("customers40.csv")
manual_customers = pd.read_csv("ManualCustomers.csv")

# get pickups/dropoffs
taxi20 = pd.read_csv("taxi_clusters_20.csv")
taxi30 = pd.read_csv("taxi_clusters_30.csv")
```

# Display Data
For: Stores unchanged, Customers with n = 20, Taxis with n = 20

```python
#Define coordinates of where we want to center our map
boulder_coords = [40.754932, -73.984016]

#Create the map
m = folium.Map(location = boulder_coords, zoom_start = 12)

#Display the map
m
```

```python
#Define the coordinates we want our markers to be at
# for stores
for i in range(len(stores)):
    #folium.Marker([stores.iloc[i][0],stores.iloc[i][1]]).add_to(my_map)
    folium.vector_layers.CircleMarker(
                    [stores.iloc[i][0],stores.iloc[i][1]],
                    radius=3,
                    tooltip = str(i),
                    color='red',
                    fill=True,
                    fill_color='red',
                    fill_opacity=0.9).add_to(m)
```

```python
# for customers

# k-means customers with n = 20
for i in range(len(customers20)):
    folium.vector_layers.CircleMarker(
                    [customers20.iloc[i][0],customers20.iloc[i][1]],
                    radius=3,
                    tooltip = str(i),
                    color='blue',
                    fill=True,
                    fill_color='blue',
                    fill_opacity=0.9).add_to(m)

# manual customers
for i in range(len(manual_customers)):
    folium.vector_layers.CircleMarker(
                    [manual_customers.iloc[i][0],manual_customers.iloc[i][1]],
                    radius=3,
                    tooltip = str(i),
                    color='black',
                    fill=True,
                    fill_color='black',
                    fill_opacity=0.9).add_to(m)
```

```python
# add pickups and dropoffs

# for 20 clusters
for i in range(len(taxi20)):
    folium.vector_layers.CircleMarker(
                    [taxi20.iloc[i][0],taxi20.iloc[i][1]],
                    radius=3,
                    tooltip = str(i),
                    color='green',
                    fill=True,
                    fill_color='green',
                    fill_opacity=0.9).add_to(m)
```

```python
# show the map
m
```

Note: Red Points = Stores, Blue Points = Customers, Black Point = Manual Customers, Green Points = Pickup/Dropoff
