# TextSearch

```python
# define query and get data from Google API
# limitation through keywords in the query (... manhattan) and radius (10200 m from the center of Manhattan)
url = "https://maps.googleapis.com/maps/api/place/textsearch/json?query=supermarket%20manhattan&location=40.7830603,-73.9712488&radius=10200&region=US&type=supermarket&key=AIzaSyBEindiZbV6koN4ycLksmKt4wBkmxwUoAU"

payload={}
headers = {}

#display results directly
response = requests.request("GET", url, headers=headers, data=payload)

#output
print(response.text)
```

## collect results as json

```python
# define empty array for results and a parameter list
supermarkets = []
params = {}

# save query results in json file
res = requests.get(url, params = params)
results =  json.loads(res.content)

# extend array
supermarkets.extend(results['results'])
#time.sleep(2)

# go through all individual results in while-loop
while "next_page_token" in results:
     params['pagetoken'] = results['next_page_token'],
     res = requests.get(url, params = params)
     results = json.loads(res.content)
     supermarkets.extend(results['results'])
     #time.sleep(2)
```

## create a dataframe

```python
# create empty arrays for longitude and latitude
shop_lat= []
shop_lon = []

# go through all results and save them to lon and lat arrays

for i in range (len(supermarkets)):
    supermarket = supermarkets[i]
    
    try:
        shop_lat.append(supermarket['geometry']['location']['lat'])
    except:
        shop_lat.append('none')
    try:
        shop_lon.append(supermarket['geometry']['location']['lng'])
    except:
        shop_lon.append('none')
```

```python
# define dataframe
df_dict= {'longitude':shop_lon,'latitude':shop_lat}

shop_df=pd.DataFrame(df_dict)

shop_df
```
